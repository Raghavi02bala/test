```python
import requests
import pandas as pd
import seaborn as sns
sns.set(rc = {'figure.figsize':(35,8)})
```

```python
lst_date = []
urls = []
df = []
```

```python
from dateutil import rrule
from datetime import datetime

a = '20200101'
b = '20201231'

for dt in rrule.rrule(rrule.DAILY,
                      dtstart=datetime.strptime(a, '%Y%m%d'),
                      until=datetime.strptime(b, '%Y%m%d')):
    lst_date.append(dt.strftime('%d-%b-%Y'))
```

```python
for i in lst_date:
  url = f'https://agmarknet.gov.in/SearchCmmMkt.aspx?Tx_Commodity=24&Tx_State=UP&Tx_District=1&Tx_Market=0&DateFrom={i}&DateTo={i}&Fr_Date={i}&To_Date={i}&Tx_Trend=0&Tx_CommodityHead=Potato&Tx_StateHead=Uttar+Pradesh&Tx_DistrictHead=Agra&Tx_MarketHead=--Select--'
  urls.append(url)
```

```python
for url in urls:
  html = requests.get(url).content
  df_list = pd.read_html(html)
  data = df_list[0]
  df.append(data)
```

```python
data = pd.concat(df, ignore_index=True)
```

```python
data = data.drop(['State Name',	'Group',	'Min Price (Rs/Quintal)',	'Max Price (Rs/Quintal)',	'Modal Price (Rs/Quintal)'], axis=1)
data.dropna(inplace=True)
data = data.set_index('Sl no.')
```
